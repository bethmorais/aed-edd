### Divisão do Dataset contemplando as **seis perguntas** mais significativas 


```
df = dataset[['Atr33', 'Atr35', 'Atr36', 'Atr39', 'Atr40', 'Atr41', 'Class']]
df.head()
```

**1o experimento:** utilizando uma divisão com 75% na base de treinamento e 25% na base de teste ou validação

```
X = df.drop('Class', axis = 1)
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```
nbc = GaussianNB()
nbc.fit(X_train, y_train)
nbc_pred = nbc.predict(X_test)
score = nbc.score(X_train, y_train)
cr = classification_report(y_test, nbc_pred)
print('Accuracy Score: {0:.3f}'.format(score))
print ('Classification Report:\n %s' % cr)
```

**2o experimento:** utilizando a média de 30 divisões com 75% na base de treinamento e 25% na base de teste ou validação

```
accuracy = []
media = 0
for i in range(30):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
  #random_state=i+1
  nbc = GaussianNB()
  nbc.fit(X_train, y_train)
  nbc_pred = nbc.predict(X_test)
  score = nbc.score(X_test, y_test)
  accuracy.append(score)
  media += score
print(media/30)
print(accuracy)
```

**3o experimento**: utilizando 5 divisões com rodízio manual, aplicando 80% na base de treinamento e 20% na base de teste ou validação

```
X_train = X[:136]
X_test = X[-34:]
y_train = y[:136]
y_test = y[-34:]
media_rodizio = 0
accuracy_rodizio = []

for i in range(5):
  nbc_rodizio = GaussianNB()
  nbc_rodizio.fit(X_train, y_train)
  nbc_rodizio_pred = nbc_rodizio.predict(X_test)
  score = nbc_rodizio.score(X_test, y_test)
  accuracy_rodizio.append(score)
  media_rodizio += score

  X_test_new = X_train[:34]
  X_train = X_train.append(X_test)
  X_train = X_train[-136:]
  X_test = X_test_new

  y_test_new = y_train[:34]
  y_train = y_train.append(y_test)
  y_train = y_train[-136:]
  y_test = y_test_new

print(media_rodizio/5)
print(accuracy_rodizio)
```

**4o experimento**: utilizando 5 divisões quebradas pelo método kFold, aplicando 80% na base de treinamento e 20% na base de teste ou validação

```
kf = KFold(n_splits=5)
i = 0
media = 0
accuracy = []

for train_index, test_index in kf.split(X):
  #print("TRAIN:", train_index, "TEST:", test_index)
  X_train, X_test = X[train_index[0]:train_index[-1]], X[test_index[0]:test_index[-1]]
  y_train, y_test = y[train_index[0]:train_index[-1]], y[test_index[0]:test_index[-1]]
  nbc = GaussianNB()
  nbc.fit(X_train, y_train)
  nbc_pred = nbc.predict(X_test)
  score = nbc.score(X_test, y_test)
  accuracy.append(score)
  media += score
  i += 1
print(media/i)
print(accuracy)
```
